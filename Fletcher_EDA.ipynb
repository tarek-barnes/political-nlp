```python
from sklearn.decomposition import NMF, TruncatedSVD

def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        
        
n_comp = 20
lsa_tfidf = TruncatedSVD(n_components=n_comp)
lsa_cv = TruncatedSVD(n_components=n_comp)
nmf_cv = NMF(n_components=n_comp)

# lsa_tfidf_data = lsa_tfidf.fit_transform(tfidf_data)
# lsa_cv_data = lsa_cv.fit_transform(cv_data)
# nmf_cv_data = nmf_cv.fit_transform(cv_data)  


# with open("lsa_tfidf_data.pickle", "wb") as f:
#     pickle.dump(lsa_tfidf_data, f)
    
# with open("lsa_cv_data.pickle", "wb") as f:
#     pickle.dump(lsa_cv_data, f)
    
# with open("nmf_cv_data.pickle", "wb") as f:
#     pickle.dump(nmf_cv_data, f)

```

## Topic modeling by publication

```python
def get_pub_dict():
    pub_dict = {}

    for pub in set(articles_train.publication):
        pub_df = articles_train[articles_train.publication == pub]
        pub_data = zip(pub_df.date, pub_df.title, pub_df.content)
        pub_dict[pub] = pub_data
        
    return pub_dict    
```

```python
pub_dict = get_pub_dict()

for pub in pub_dict:
    print("{} \n \n".format(pub.upper()))
#     titles = [k[1] for k in pub_dict[pub]]
    articles = [k[2] for k in pub_dict[pub]]
    
#     cv_titles_data = count_vectorizer.fit_transform(titles)
#     tfidf_titles_data = tfidf_vectorizer.fit_transform(titles)
    
    cv_articles_data = count_vectorizer.fit_transform(articles)
    tfidf_articles_data = tfidf_vectorizer.fit_transform(articles)
    
    lsa_tfidf_articles_data = lsa_tfidf.fit_transform(tfidf_articles_data)
    lsa_cv_articles_data = lsa_cv.fit_transform(cv_articles_data)
    nmf_cv_articles_data = nmf_cv.fit_transform(cv_articles_data)
    
    print("LSA TDIDF:")
    print(display_topics(lsa_tfidf, tfidf_vectorizer.get_feature_names(),10))
    
    print("LSA CV:")
    print(display_topics(lsa_cv, count_vectorizer.get_feature_names(),10))
    
    print("NMF CV:")
    print(display_topics(nmf_cv, count_vectorizer.get_feature_names(),10))    
```
